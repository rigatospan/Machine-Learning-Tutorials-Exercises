## Visualizing Data

Plot the function $f(x) = x^2$ in the interval $[-5,5]$

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x**2
```

```python
a = 5
# generate 100 points in the interval [-a, a]
x=np.linspace(-5, 5,100)
y=f(x)
 
# plt.figure(figsize=(5,5))
plt.plot(x,y)
plt.xlabel('x ')
plt.ylabel('$y$')
plt.title('$f(x)=x^2$')
plt.show()
```

```python
# Now use object oriented (OO) way

x = np.linspace(-a, a, 100)

# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, f(x), label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend.
```

Now create a sample of points drawn from a multi-Gaussian distirbution with mean $[1,0.5]$
and covariance matrix $\text{diag}(0.4,0.4)$

```python
#draw points from a 2-D Gaussian distribution
#Use transpose to set it in 2x100 form
mean = np.array([1, 0.5])
covariance = np.array([[0.4, 0], [0, 0.4]])
n_points = 100

points = np.random.multivariate_normal(mean, covariance, n_points)
print(f'points have the shape {points.shape}')
plt.scatter(points.T[0], points.T[1])
plt.show()
```

Now we also plot the contour lines of the 2-dim Gaussian, which are circles of the form
$$ z = \frac{1}{\sigma_x^2} (x-x_{\text{center}})^2 + \frac{1}{\sigma_y^2} (y-y_{\text{center}})^2 $$

```python
plt.scatter(points.T[0], points.T[1])
plt.plot(1,0.5,'rx')
# contour plot of 2D Gaussian are circles
# to use contour plot we should create a meshgrid
x=np.arange(-2,3,0.1)
y=np.arange(-2,3,0.1)
xv,yv=np.meshgrid(x,y)
z=1/0.4**2*(xv-1)**2+1/0.4**2*(yv-0.5)**2
z_values = [1, 2, 5, 10, 20 ]
plt.contour(xv, yv, z, z_values)
plt.axis('equal')
plt.show()
```
